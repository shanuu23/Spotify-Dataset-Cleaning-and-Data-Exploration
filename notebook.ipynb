##### Only two modes are there so "category" type datatype will be easy to deal with instead of "int64" 

```python
# converting into category
df['mode'] = df['mode'].astype('category')
```

```python
# Creating a backup in case we mess up with the dataset
df_backup = df.reset_index().sort_values('track_popularity',ascending=False).set_index('track_id')
```

```python
# Sorting by track popularity and dropping duplicates 
df = df.reset_index().sort_values('track_popularity', ascending=False).drop_duplicates('track_id', keep='first').set_index('track_id')
```

##### Here, all duplicates are dropped—meaning if a track appears in one playlist, it won't be included in another. This will be useful later when we visualize the top song of the top genre and the top genre of each playlist, ensuring the same song doesn't appear in more than one playlist.

### EDA

#### Here, we're identifying the albums with the highest number of hit tracks in the dataset, based on quantity.

```python
album_popularity = df.groupby('track_album_name').size().reset_index().rename(columns={0:'no_of_tracks'})
album_popularity = album_popularity.sort_values(by='no_of_tracks',ascending=False).reset_index(drop=True)
top_album = album_popularity[0:5]
sns.barplot(x='no_of_tracks',y='track_album_name',data=top_album,color='red')
plt.title('Album with most number of hit tracks')
plt.show()
```

#### Here, we are retrieving the top song of the top subgenre, and the top subgenre of the top genre for each playlist, along with the artist names and track popularity scores.

```python
top_playlists = df.groupby('playlist_name')['track_popularity'].mean().sort_values(ascending=False).head(5).index

results = []

for playlist in top_playlists:
    df_playlist = df[df['playlist_name'] == playlist]
    top_genre = df_playlist.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False).idxmax()
    df_genre = df_playlist[df_playlist['playlist_genre'] == top_genre]
    top_subgenre = df_genre.groupby('playlist_subgenre')['track_popularity'].mean().sort_values(ascending=False).idxmax()
    df_subgenre = df_genre[df_genre['playlist_subgenre'] == top_subgenre]
    top_track_row = df_subgenre.sort_values(by='track_popularity', ascending=False).iloc[0]

    results.append({
        'playlist_name': playlist,
        'top_genre': top_genre,
        'top_subgenre': top_subgenre,
        'track_name': top_track_row['track_name'],
        'track_artist': top_track_row['track_artist'],
        'track_popularity': top_track_row['track_popularity']
    })

result_df = pd.DataFrame(results)
result_df
```

#### Here, we are visualizing the top song of the top genre, and the top genre for each playlist, along with the artist names and track popularity scores.

```python
custom_order = [
    "Global Top 50",
    "Today's Top Hits",
    "Top Gaming Tracks",
    "Viva Latino",
    "Soft Pop Hits"
]

result_df['playlist_name'] = pd.Categorical(result_df['playlist_name'], categories=custom_order, ordered=True)
result_df_sorted = result_df.sort_values('playlist_name')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=result_df_sorted,
    x='track_popularity',
    y='playlist_name',
    hue='top_genre',
    dodge=False,
    palette='Set2'
)

for index, row in result_df_sorted.iterrows():
    plt.text(row['track_popularity'] + 1, index, f"{row['track_name']} ({row['track_artist']})", va='center')

plt.title("Top Tracks from Top 5 Playlists by Popularity")
plt.xlabel("Track Popularity")
plt.ylabel("Playlist Name")
plt.legend(title='Top Genre', bbox_to_anchor=(1,1), loc='lower right')
plt.tight_layout()
plt.show()
```
